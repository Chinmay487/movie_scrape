```python
def get_budget_info(year):
    df_year = df_copy[df_copy['release_year'] == year]
    max_budget = df_year['budget(in million)'].max()
    min_budget = df_year['budget(in million)'].min()
    average_budget = df_year['budget(in million)'].mean()
    total_amount_spent = df_year['budget(in million)'].sum()
    movie_with_max_budget = df_year[df_year['budget(in million)'] == max_budget]['title'].to_string(index=False)
    movie_with_min_budget = df_year[df_year['budget(in million)'] == min_budget]['title'].to_string(index=False)
#     print(f'max budget is {max_budget} | movie is {movie_with_max_budget}')
#     print(f'min budget is {min_budget} | movie is {movie_with_min_budget}')
    return max_budget,min_budget,movie_with_max_budget,movie_with_min_budget,average_budget,total_amount_spent
```

```python
def get_revenue_info(year):
    df_year = df_copy[df_copy['release_year'] == year]
    max_revenue = df_year['revenue(in million)'].max()
    min_revenue = df_year['revenue(in million)'].min()
    average_revenue = df_year['revenue(in million)'].mean()
    total_revenue_generated = df_year['revenue(in million)'].sum()
    movie_with_max_revenue = df_year[df_year['revenue(in million)'] == max_revenue]['title'].to_string(index=False)
    movie_with_min_revenue = df_year[df_year['revenue(in million)'] == min_revenue]['title'].to_string(index=False)
#     print(f'max revenue is {max_revenue} | movie is {movie_with_max_revenue}')
#     print(f'min revenue is {min_revenue} | movie is {movie_with_min_revenue}')
    return max_revenue,min_revenue,movie_with_max_revenue,movie_with_min_revenue,average_revenue,total_revenue_generated
```

```python
def get_df(years):
    max_budget_list = list()
    min_budget_list = list()
    movie_with_max_budget_list = list()
    movie_with_min_budget_list = list()
    max_revenue_list = list()
    min_revenue_list = list()
    movie_with_max_revenue_list = list()
    movie_with_min_revenue_list = list()
    average_budget_list = list()
    total_amount_list = list()
    average_revenue_list = list()
    total_revenue_list = list()
    
    for year in years:
        max_budget,min_budget,movie_with_max_budget,movie_with_min_budget,average_budget,total_amount_spent = get_budget_info(year)
        max_revenue,min_revenue,movie_with_max_revenue,movie_with_min_revenue,average_revenue,total_revenue_generated = get_revenue_info(year)
        max_budget_list.append(max_budget)
        min_budget_list.append(min_budget)
        movie_with_max_budget_list.append(movie_with_max_budget)
        movie_with_min_budget_list.append(movie_with_min_budget) 
        max_revenue_list.append(max_revenue)
        min_revenue_list.append(min_revenue)
        movie_with_max_revenue_list.append(movie_with_max_revenue)
        movie_with_min_revenue_list.append(movie_with_min_revenue)
        average_budget_list.append(average_budget)
        total_amount_list.append(total_amount_spent)
        average_revenue_list.append(average_revenue)
        total_revenue_list.append(total_revenue_generated)
        
    
    info_dict = {
        'year' : years,
        'max_budget' : max_budget_list,
        'movie_with_max_budget' : movie_with_max_budget_list,
        'min_budget' : min_budget_list,
        'movie_with_min_budget' : movie_with_min_budget_list,
        'max_revenue' : max_revenue_list,
        'movie_with_max_revenue' : movie_with_max_revenue_list,
        'min_revenue' : min_revenue_list,
        'movie_with_min_revenue' : movie_with_min_revenue_list,
        'average_budget': average_budget_list,
        'total_amount_spent': total_amount_list,
        'average_revenue': average_revenue_list,
        'total_revenue_generated': total_revenue_list
    }
    
    return pd.DataFrame(info_dict)
```
